## Processing outliers
Most station usage statistics will be in regards to user-station interaction. In the case of prolonged rides (possible theft after usage) we'll maintain this data. It will still count as a user-station interaction. 

We will remove rides that are too short. This will be done using the following logic: 12 mph is considered a low to moderate speed for bike rides. Most NiceRide bikes will be taken at this speed or slower due to the mechanics of these bikes. Average block size is 310-325 ft. At 12mph 63,360 feet per hour. A person could travel ~3 city blocks per minute. If a person were to travel to a different station (at least a city block away) a low to moderate time to due so would be ~20 seconds. Most rides should be longer than 20 seconds. Rides that are shorter are most likely people returning a bike back into the same rack as it was removed to grab a different bike. A possible reason for this is that the bike may be damaged, or it could be a NiceRide employee testing docking stations to ensure they work. We don't want to use this data in user-station interactions because it may cause inflated use cases. To be on a safer side of things we will remove all rides that are shorter than ~10 seconds.

```python
for x in range(8) :
    print('201'+str(x))
    print(df_list[x]['duration'].count())
```

```python
for x in range(8) :
    df_list[x] = df_list[x][df_list[x]['duration'] > 10]
```
