## This notebook will be used for calculating ride patterns of stations
Using the bokeh plots as a visual reference. This notebook is built to be a back end calculator for ride aggregation and pattern numerical description.

```python
import pandas as pd
import numpy as np
from collections import defaultdict
```

```python
ridership_dict = defaultdict()
station_dict = defaultdict()

for year in [2010 + x for x in range(8)] :

    ridership_dict[year] = pd.read_csv("~/NiceRideMN/Nice_Ride_data/"+str(year)+"/NiceRide_trip_history_"+str(year)+".csv")
    station_dict[year] = pd.read_csv("~/NiceRideMN/Nice_Ride_data/"+str(year)+"/NiceRide_station_"+str(year)+".csv")
    
NR_ridership = pd.concat(ridership_dict)
NR_station = pd.concat(station_dict)
```

```python
summit_17 = [30162,30155,30085,30086,30099,30101,30106,30103,30137] 
# Station 30084 was removed in 2017 and 30155 moved south to where 30084 was

summit_16 = [30162,30155,30085,30086,30099,30101,30106,30103,30137, 30084]
hiawatha = [30159,30153,30174,30149, 30148, 30150] # This grouping is the same in 2017 as 2016
```

```python
nr_17 = NR_ridership.loc[2017,:]
nr_16 = NR_ridership.loc[2016,:]
```

```python
# building new dataframes based on hiawatha corridor
# format is untenable for larger datasets and would require optimization

nr_hia_17 = pd.DataFrame()
nr_hia_16 = pd.DataFrame()

for x in hiawatha :
    
    temp_17 = nr_17[nr_17.Start_id == x]
    temp_16 = nr_16[nr_16.Start_id == x]
    
    nr_hia_17 = nr_hia_17.append(temp_17)
    nr_hia_16 = nr_hia_16.append(temp_16)
    
nr_hia_17 = nr_hia_17.drop(['Start_name', 'End_name'], axis=1).reset_index(drop=True)
nr_hia_16 = nr_hia_16.drop(['Start_name', 'End_name'], axis=1).reset_index(drop=True)
```

```python
nr_hia_17['counts'] = 0 # create a new column filled with zeroes for a default value

# fill the count column by count aggregrating terminals by start_id and end_id
ride_counts = nr_hia_17.groupby(by=['Start_id', 'End_id'],axis=0, as_index=False)['counts'].count()
```
